```python
df1= pd.read_excel('Excel_old.xlsx')
df2= pd.read_excel('Excel_new.xlsx')
```

```python
cmp=df1.values==df2.values
```

```python
cmp
```

```python
import numpy as np
```

```python
rows,cols= np.where(cmp==False)
```

```python
cmp
```

```python
for item in zip(rows,cols):
        df1.iloc[item[0],item[1]]=df1.iloc[item[0], item[1]]-df2.iloc[item[0], item[1]] 
```

```python
df1
```

```python
dfDiff = df1.copy()
for row in range(dfDiff.shape[0]):
    for col in range(dfDiff.shape[1]):
        value_OLD = df1.iloc[row,col]
        value_NEW = df2.iloc[row,col]
        if value_OLD==value_NEW:
            dfDiff.iloc[row,col] = 0
        else:
            dfDiff.iloc[row,col] = value_OLD-value_NEW
```
